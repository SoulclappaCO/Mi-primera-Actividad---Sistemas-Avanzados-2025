```python
# ======================================================
# RED NEURONAL INDUSTRIAL - GRICOL S.A.
# Autor: ChatGPT + Yerson Gómez
# Representa los procesos y subprocesos industriales interconectados
# ======================================================

import matplotlib.pyplot as plt
import networkx as nx

# --- Procesos principales ---
procesos = {
    "Fundición": ["Coquilla", "Corazones", "Estampado", "Hermeticidad", "Ajax"],
    "Recubrimientos": ["Plástico", "Plástico2", "Metal", "Pintura", "Especiales", "Vibrado", "Roscado Mecánico"],
    "Mecanizado": ["Tornos Automáticos", "Transfer", "Válvulas", "CNC", "Revólver y Tubo"],
    "Inyección": ["ABS", "Polipropileno", "Acetal", "Caucho"],
    "Soldadura": ["Troquelado", "Soldadura"],
    "Ensamble": ["Ensamble", "Válvulas", "Termoformado"],
    "Pulido": ["Pulido"]
}

# Crear grafo
G = nx.Graph()

# Nodo central y salida final
nodo_central = "Núcleo de Manufactura\nGRICOL"
salida_final = "Válvula PUSH DUCHS\n(Resultado Final)"

# Agregar conexiones de procesos
for proceso, subprocesos in procesos.items():
    G.add_edge(proceso, nodo_central)
    for sub in subprocesos:
        G.add_edge(proceso, sub)

# Conectar todos los procesos entre sí (red totalmente unida)
procesos_lista = list(procesos.keys())
for i in range(len(procesos_lista)):
    for j in range(i+1, len(procesos_lista)):
        G.add_edge(procesos_lista[i], procesos_lista[j])

# Conectar salida final
G.add_edge(nodo_central, salida_final)

# Posiciones manuales para apariencia tipo “enredada” (similar a tu dibujo)
pos = nx.spring_layout(G, k=0.8, seed=15)

plt.figure(figsize=(15, 10))
plt.title("RED NEURONAL INDUSTRIAL - GRICOL S.A.", fontsize=16, fontweight='bold')

# --- Dibujar nodos ---
nx.draw_networkx_nodes(G, pos,
                       nodelist=list(procesos.keys()),
                       node_color="#1f77b4", node_size=2000, alpha=0.85)

nx.draw_networkx_nodes(G, pos,
                       nodelist=[nodo_central],
                       node_color="#9467bd", node_size=2500, alpha=0.9)

nx.draw_networkx_nodes(G, pos,
                       nodelist=[salida_final],
                       node_color="#2ca02c", node_size=2500, alpha=0.9)

# Subprocesos (nodos pequeños)
subprocesos = [s for subs in procesos.values() for s in subs]
nx.draw_networkx_nodes(G, pos,
                       nodelist=subprocesos,
                       node_color="#ff7f0e", node_size=700, alpha=0.8)

# --- Dibujar aristas ---
nx.draw_networkx_edges(G, pos, width=1.8, alpha=0.5)

# --- Etiquetas ---
nx.draw_networkx_labels(G, pos, font_size=9, font_color="black", font_weight="bold")

plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
# ======================================================
# RED NEURONAL INDUSTRIAL TOTAL - GRICOL S.A.
# Versión completamente interconectada
# ======================================================

import matplotlib.pyplot as plt
import networkx as nx

# --- Procesos principales y subprocesos ---
procesos = {
    "Fundición": ["Coquilla", "Corazones", "Estampado", "Hermeticidad", "Ajax"],
    "Recubrimientos": ["Plástico", "Plástico2", "Metal", "Pintura", "Especiales", "Vibrado", "Roscado Mecánico"],
    "Mecanizado": ["Tornos Automáticos", "Transfer", "Válvulas", "CNC", "Revólver y Tubo"],
    "Inyección": ["ABS", "Polipropileno", "Acetal", "Caucho"],
    "Soldadura": ["Troquelado", "Soldadura"],
    "Ensamble": ["Ensamble", "Válvulas", "Termoformado"],
    "Pulido": ["Pulido"]
}

# Crear grafo
G = nx.Graph()

# Nodos centrales
nucleo = "Núcleo de Manufactura\nGRICOL"
salida = "Válvula PUSH DUCHS\n(Resultado Final)"

# --- Añadir conexiones básicas ---
for proceso, subprocesos in procesos.items():
    G.add_edge(proceso, nucleo)
    for sub in subprocesos:
        G.add_edge(proceso, sub)

# --- Conectar todos los procesos entre sí ---
procesos_lista = list(procesos.keys())
for i in range(len(procesos_lista)):
    for j in range(i+1, len(procesos_lista)):
        G.add_edge(procesos_lista[i], procesos_lista[j])

# --- Conectar todos los subprocesos entre sí ---
subprocesos = [s for subs in procesos.values() for s in subs]
for i in range(len(subprocesos)):
    for j in range(i+1, len(subprocesos)):
        G.add_edge(subprocesos[i], subprocesos[j])

# --- Conectar salida final ---
G.add_edge(nucleo, salida)

# --- Layout visual ---
pos = nx.spring_layout(G, k=0.4, iterations=60, seed=4)

plt.figure(figsize=(16, 12))
plt.title("RED NEURONAL INDUSTRIAL TOTAL - GRICOL S.A.", fontsize=16, fontweight='bold')

# Nodos por categoría
nx.draw_networkx_nodes(G, pos, nodelist=procesos_lista, node_color="#1f77b4", node_size=2000, alpha=0.85, label="Procesos")
nx.draw_networkx_nodes(G, pos, nodelist=subprocesos, node_color="#ff7f0e", node_size=700, alpha=0.8, label="Subprocesos")
nx.draw_networkx_nodes(G, pos, nodelist=[nucleo], node_color="#9467bd", node_size=2500, alpha=0.9, label="Núcleo Central")
nx.draw_networkx_nodes(G, pos, nodelist=[salida], node_color="#2ca02c", node_size=2500, alpha=0.9, label="Salida Final")

# Aristas (todas las conexiones)
nx.draw_networkx_edges(G, pos, width=0.6, alpha=0.3)

# Etiquetas de los nodos
nx.draw_networkx_labels(G, pos, font_size=8.5, font_color="black", font_weight="bold")

# Configuración final
plt.legend(scatterpoints=1, fontsize=10, loc="upper left")
plt.axis('off')
plt.tight_layout()
plt.show()
```
